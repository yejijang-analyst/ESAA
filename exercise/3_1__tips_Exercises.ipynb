### Step 1. Import the necessary libraries:

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
sns.set_style("white") # 가독성 스타일
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Tips/tips.csv).

### Step 3. Assign it to a variable called tips

```python
url = "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Tips/tips.csv"
tips = pd.read_csv(url)
tips
```

### Step 4. Delete the Unnamed 0 column

```python
tips_drop_df = tips.drop('Unnamed: 0', axis=1)
tips_drop_df.head(3)
```

```python
tips.columns
```

### Step 5. Plot the total_bill column histogram

```python
ax = plt.subplot()
ax = sns.distplot(tips['total_bill'])

# take out the right and upper borders
sns.despine()
```

### Step 6. Create a scatter plot presenting the relationship between total_bill and tip

```python
## 내풀이 오답
ax = plt.subplot()
ax = sns.regplot(x='total_bill', y='tip', data = tips)
```

```python
sns.jointplot(x ="total_bill", y ="tip", data = tips)
```

### Step 7.  Create one image with the relationship of total_bill, tip and size.
#### Hint: It is just one function.

```python
# 내풀이 오답
scatter = sns.lmplot(x='total_bill', y='tip', data=tips, hue='size', fit_reg=False)
plt.show()
```

```python
sns.pairplot(tips)
```

### Step 8. Present the relationship between days and total_bill value

```python
ax = plt.subplots()
ax = sns.barplot(x='day', y='total_bill', data=tips)
```

```python
# 다른 풀이 솔루션
#jitter: 데이터 포인트를 겹치지 않도록 약간의 무작위 더미값을 추가하여 시각화합니다.
#이것을 True로 설정하면 데이터 포인트가 약간 무작위로 분산되어 보여집니다.
#sns.stripplot() 함수는 Seaborn 라이브러리를 사용하여 만들어진 함수로, 데이터의 분포를 보여주는 데 사용됩니다.
#주로 범주형 변수와 연속형 변수 간의 관계를 시각화할 때 유용합니다.

sns.stripplot(x = "day", y = "total_bill", data = tips, jitter = True);
```

### Step 9. Create a scatter plot with the day as the y-axis and tip as the x-axis, differ the dots by sex

```python
## 내풀이 오답 : lm 함수는 회귀선 그리는 함수로 xy 에는 연속형 변수 필요
scatter = sns.lmplot(x='day', y='tip', data=tips, hue='sex', fit_reg=False)
plt.show()
```

```python
sns.stripplot(x = "tip", y = "day", hue = "sex", data = tips, jitter = True);
```

### Step 10.  Create a box plot presenting the total_bill per day differetiation the time (Dinner or Lunch)

```python
ax = plt.subplots()
ax = sns.boxplot(x='day', y='total_bill', hue ='time',  data=tips)
ax.set_title('Boxplot of total bill by time of day')
```

### Step 11. Create two histograms of the tip value based for Dinner and Lunch. They must be side by side.

```python
# sns.set(style="ticks")는 이 부분은 Seaborn에서 그래프의 스타일을 설정하는 부분입니다.
# "ticks" 스타일은 그래프의 눈금에 작은 눈금 표시를 사용하는 스타일을 지정합니다.
```

```python
sns.set(style = "ticks")
g = sns.FacetGrid(tips, col="time") # "time" 열을 기준으로 나누어진 두 개의 하위 그래프가 생성
g.map(plt.hist, "tip") #각 하위 그래프에는 "tip" 열 데이터의 히스토그램
```

### Step 12. Create two scatterplots graphs, one for Male and another for Female, presenting the total_bill value and tip relationship, differing by smoker or no smoker
### They must be side by side.

```python
# 실행속도 엄청 느림. 매핑 펑션에는 sns 말고 plt 쓰는게..
gg = sns.FacetGrid(tips, col = "sex")
gg.map(sns.stripplot, x='total_bill', y='tip',hue='smoker', data=tips,jitter = True )
plt.legend(title='Smoker', labels=['Non-Smoker', 'Smoker'])
```

```python
g = sns.FacetGrid(tips, col = "sex", hue = "smoker")
g.map(plt.scatter, "total_bill", "tip", alpha =.7) #  alpha 매개변수로 투명도를 설정합니다.

g.add_legend();
#FacetGrid 객체에 add_legend 메서드를 호출하여 범례를 추가합니다.
```

```python
g = sns.FacetGrid(tips, col="sex", hue="smoker") # 무조건 hue 가 facet 에 와야됨.
g.map(sns.scatterplot, "total_bill", "tip", alpha=0.7)
g.add_legend()
```
