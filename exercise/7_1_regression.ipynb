# **1. 단순 선형 회귀 분석**
- 전복의 나이를 예측하는 선형회귀모델을 생성하세요.
- 전복의 ‘성별’, ‘키’, ‘지름’, ‘높이’, ‘전체무게’, ‘몸통무게’, ‘내장무게’, ‘껍질무게’를 이용해 ‘껍질의 고리 수’를 예측한 뒤, **예측된 ‘껍질의 고리 수’에 1.5를 더하면 전복의 나이**가 됩니다.

```python
# 기본 모듈 불러오기
import numpy as np
import pandas as pd
```

**1) 데이터 load 및 변형**

```python
# 데이터 로드
data = pd.read_csv("/content/drive/MyDrive/commit/commit_github_esaa/data/abalone (2).csv")
data.head()


```

```python
print(data.shape)
```

```python
# 성별 M은 Male, F는 Female, I는 Infant 이므로 따로 열 만들기
for label in "MFI":
    data[label] = data["Sex"] == label # 논리연산자를 우선연산자로 취급
data.drop('Sex', axis=1, inplace=True)
```

```python
data.head()
```

**2) X, y 선택**
: y는 Rings열, X는 Rings열을 제외한 나머지를 선택하되 전부 실수가 되도록 한다.

```python
# X,y 데이터 선택
y = data["Rings"]

X = data.drop("Rings", axis=1)
X = X.apply(pd.to_numeric, errors='coerce').astype(float)



y.head()
```

```python
X.head()
```

 **3) train/test set 분리**

```python
# 필요한 모듈 불러오기
from sklearn.model_selection import train_test_split
```

```python
# train과 test set 분리 (train:test = 7:3 비율로)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=2)
```

**4) 선형회귀모델 생성, 모델 예측치 구하기**

```python
#필요한 모듈 불러오기
from sklearn.linear_model import LinearRegression
```

```python
#선형회귀모델 생성 및 훈련
lr = LinearRegression()
lr.fit(X_train, y_train)
y_preds = lr.predict(X_test)
```

```python
# 모델 예측치 구하기

# 모델 예측치를 활용해 최종적으로 전복의 나이를 예측
age = 1.5 + y_preds
age
```

**5) 모델 평가: MSE, RMSE, R2 score, corr 구하기**

```python
#필요한 모듈 불러오기
from sklearn.metrics import mean_squared_error, r2_score
```

- MSE, RMSE

```python
#mse, rmse
mse = mean_squared_error(y_test,y_preds)

rmse = np.sqrt(mse)

print('MSE : {0:.3f} , RMSE : {1:.3F}'.format(mse , rmse))
```

- R2 score

```python
#R2 score 측정
r2_score(y_test, y_preds)
```

- 회귀 절편값

```python
#회귀 절편 값
print('절편 값:',lr.intercept_)
```

- 회귀 계수 값

```python
#회귀 계수 값
print('회귀 계수값:', np.round(lr.coef_, 1))
```

- 상관계수

Hint: corr 함수 이용.

```python
# 상관계수 구하기
X_y = pd.concadt([X_train, y_train], axis=1)
correlation = X_y.corr()
correlation
```
