### Step 4. Discover what is the mean age per occupation

```python
users.groupby("occupation")['age'].mean()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
genders = users.groupby('occupation').gender.value_counts(normalize= True)*100
genders.head()

# "occupation"과 "gender" 열을 인덱스로 하고
# Series 를 반환했으므로 무조건 아래에서 DateFrame 시켜줘야 함.
```

```python
col_percents = pd.DataFrame(genders)
col_percents.rename(columns={'gender': 'percent'}, inplace=True)
print(col_percents.head())

```

```python
# groupby로 인해 인덱스화 된 "occupation"과 "gender" 열 말고 새로운 인덱스 부르기
new_ind = col_percents.reset_index(level=['occupation', 'gender'])
print(new_ind.head())
```

```python
males = new_ind[new_ind['gender'] == 'M'].sort_values(by='percent', ascending=False)
print(males)
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
users.groupby("occupation").age.agg([min, max])
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
users.groupby(['occupation', 'gender']).age.agg('mean')
```
