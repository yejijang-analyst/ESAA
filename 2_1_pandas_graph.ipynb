```python
tips = sns.load_dataset("tips")

print(tips.head())
print(type(tips))
```

```python
fig = plt.figure()
axes1 = fig.add_subplot(1, 1, 1)
```

```python
axes1.hist(tips['total_bill'], bins=10) # 히스토그램 : 변수 하나를 사용하는 일변량 그래프

axes1.set_title('Histogram of Total Bill')
axes1.set_xlabel('Frequency')
axes1.set_ylabel('Total Bill')

fig
```

```python
# 변수 2개를 사용하는 이변량 그래프
#scatter 메서드 활용
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1, 1, 1)
axes1.scatter(tips['total_bill'], tips['tip'])
axes1.set_title('Scatterplot of Total Bill vs Tip')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

```python
# 박스그래프 : 이산형 변수와 연속형 변수를 함께 사용.
# boxplot: 데이터 프레임에서 남인 데이터와 여인 데이터에서 tip 열데이터만 추출하여 리스트에 담아 전달. 그리고 라벨 붙이기.

boxplot = plt.figure()
axes1 = boxplot.add_subplot(1,1,1)

axes1.boxplot([tips[tips['sex'] == 'Female']['tip'],tips[tips['sex'] == 'Male']['tip'] ], labels = ['Female', 'Male'])

axes1.set_xlabel('Sex')
axes1.set_ylabel('Tip')
axes1.set_title('Boxplot of Tips by Sex')
```

```python
#  다변량 그래프
# 남 여 와 같은 문자열은 산점도 그래프의 색상을 지정하는 값으로 사용 불가
# 문자열을 정수로 치환하는 함수를 만들어야함

def record_sex(sex):
  if sex == 'Female':
        return 0
  else:
        return 1
```

```python
# 브로드캐스팅 함수 10장.
# 문자열을 숫자로 반환한 값을 데이터 프레임에 sex_color   열을 새로 생성하여 추가

tips['sex_color'] = tips['sex'].apply(record_sex) # sex 열에 record sex 함수를 브로드 캐스팅함.
print(tips.head())
```

```python
# s는 점의 크기를 c 는 점의 색상을 의미.
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1, 1, 1)
axes1.scatter(
    x=tips['total_bill'],
    y=tips['tip'],
    s=tips['size'] * 10,
    c=tips['sex_color'],
    alpha=0.5)

axes1.set_title('Total Bill vs Tip Colored by Sex and Sized by Size')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

```python
import seaborn as sns
tips = sns.load_dataset("tips")
```

```python
#seaborn 라이브러리로 히스토그램을 그리려면 subplots, distplot 메서드를 사용

ax = plt.subplots()
ax = sns.distplot(tips['total_bill'])
ax.set_title('Total Bill Histogram with Density Plot')
```

```python
# 밀집도 그래프를 제외하려면 kde = False 설정
ax = plt.subplots() # 기본틀 및 격자 생성
ax = sns.distplot(tips['total_bill'],kde=False) # 생성한 격자 그래프에 플랏 그리기
ax.set_title('Total Bill Histogram')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Frequency')
```

```python
# 히스토그램 제외
ax = plt.subplots() # 기본틀 및 격자 생성
ax = sns.distplot(tips['total_bill'],hist=False) # 생성한 격자 그래프에 플랏 그리기
ax.set_title('Total Bill Histogram')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Frequency')
```

```python
# 여러 그래프를 한 번에 출력
# 히스토그램 + 밀집도 + rug 그래프

ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], rug = 1) #  x에 열 데이터를 지정하는 대신 data에 데이터프레임 전체를 넘겨야 합니다.
ax.set_title('Total Bill Histogram with Density and Rug Plot')
ax.set_xlabel('Total Bill')
```

```python
#count 그래프 : 이산값을 나타낸 그래프
ax = plt.subplots()
ax = sns.countplot('day', data=tips)
ax.set_title('Count of days')
ax.set_xlabel('Day of the Week')
ax.set_ylabel('Frequency')
```

```python
ax = plt.subplots()
ax = sns.countplot(x = 'day', data=tips)  # 'x' 인자에 'day'를 지정
ax.set_title('Count of days')
ax.set_xlabel('Day of the Week')
ax.set_ylabel('Frequency')
```

```python
# regplot 산점도와 회귀선 같이 그리기
ax = plt.subplots()
ax = sns.regplot(x='total_bill', y='tip', data=tips)
ax.set_title('Scatterplot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```python
# 회귀직선 없애기
ax = plt.subplots() # 만약 하나의 그래프만을 그린다면 ax = plt.subplots()를 사용하지 않아도 가능
ax = sns.regplot(x='total_bill', y='tip', data=tips, fit_reg=0)  # 나중에 데이터 이름을 언급해주는 방식
ax.set_title('Scatterplot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```python
#jointplot 메서드를 통해 산점도와 히스토그래을 한 번에 그릴 수 있음
joint = sns.jointplot(x='total_bill', y='tip', data=tips)
joint.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
joint.fig.suptitle('Joint Plot of Total Bill and Tip', fontsize=10, y=1.03)
```

```python
# 이차원 육각형으로 표현
joint = sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
joint.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
joint.fig.suptitle('Joint Plot of Total Bill and Tip', fontsize=10, y=1.03)
```

```python
# shade =1 그래프에 음영효과 줄 수 있음
# kdeplot 메서드를 활용하여 이차원 밀집도를 만들 수 있다.
ax = plt.subplots()
ax = sns.kdeplot(x=tips['total_bill'],
                 y=tips['tip'],
                 shade=True)
ax.set_title('Kernel Density Plot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```python
sns.set(style="whitegrid")
g = sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde', shade=True)
g.set_axis_labels('Total Bill', 'Tip')
g.fig.suptitle('Kernel Density Plot of Total Bill and Tip', fontsize=12, y=1.03)

```

```python
# 바그래프로 지정한 변수의 평균을 계산하여 그릴 수 있음
ax = plt.subplots()
ax = sns.barplot(x='time', y='total_bill', data=tips)
ax.set_title('Bar plot of average total bill for time of day')
ax.set_xlabel('Time of day')
ax.set_ylabel('Average total bill')
```

```python
# 박스 그래프는 sns 를 사용하는 것이 이득이다..
ax = plt.subplots()
ax = sns.boxplot(x='time', y='total_bill', data=tips)
ax.set_title('Boxplot of total bill by time of day')
ax.set_xlabel('Time of day')
ax.set_ylabel('Total Bill')
```

```python
# 모호한 데이터 분산을 확인하기 위해 바이올린 그래프 사용
ax = plt.subplots()
ax = sns.violinplot(x='time', y='total_bill', data=tips)
ax.set_title('Violin plot of total bill by time of day')
ax.set_xlabel('Time of day')
ax.set_ylabel('Total Bill')
```

```python
fig = sns.pairplot(tips)
```

```python
#위의 대각선을 기준으로 위아래 중복 그래프 수정
# map_upper:  대각전을 기준으로 위쪽에 그릴 그래프 지정
# map_lower:  대각전을 기준으로 아래쪽에 그릴 그래프 지정
# map_diag:  대각전을 중심으로 그릴 그래프 지정
pair_grid = sns.PairGrid(tips)
pair_grid = pair_grid.map_upper(sns.regplot)
pair_grid = pair_grid.map_lower(sns.kdeplot)
pair_grid = pair_grid.map_diag(sns.distplot, rug=True)
plt.show()

#간단한 쌍별 플롯을 생성하고 스타일을 변경하려면 sns.pairplot(tips)을 사용하는 것이 편리하며,
#더 복잡한 사용 사례나 개별적인 서브플롯 제어가 필요한 경우 sns.PairGrid(tips)를 사용할 수 있습니다.
```

```python
# seaborn 라이브러리로  다변량 그래프 그리기
# 색상 추가 : hue = 'sex'  hue 인자값으로 쉽게 가능.
# hue='sex'를 사용하여 성별에 따라 바이올린 플롯을 색으로 구분합니다.
# split=True를 설정하면 각 바이올린 안에 여성과 남성 데이터가 분리되어 표시됩니다.
violin, ax = plt.subplots()
ax = sns.violinplot(x='time', y='total_bill', hue='sex', data=tips, split=True)
plt.show()
```

```python
#lmplot: lmplot은 주로 데이터의 산점도와 선형 회귀 선을 함께 그리는 데 사용됩니다.
# -  여러 그룹 또는 범주형 변수에 따른 선형 회귀를 비교할 때 유용합니다.
# regplot: regplot은 단일 산점도와 해당 선형 회귀 선을 그리는 데 사용됩니다.
# -  단일 그래프의 경우 주로 regplot을 사용합니다.

scatter = sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', fit_reg=False)
plt.show()
```

```python
fig = plt.figure()
seaborn_styles = ['darkgrid', 'whitegrid', 'dark', 'white', 'ticks']
for idx, style in enumerate(seaborn_styles):
    plot_position = idx + 1
    with sns.axes_style(style):
        ax = fig.add_subplot(2, 3, plot_position)
        violin = sns.violinplot(x='time', y='total_bill', data=tips, ax=ax)
        violin.set_title(style)

fig.tight_layout()
```
